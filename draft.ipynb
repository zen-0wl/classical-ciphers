```python
# library process of monoalphabetic cipher
import sys
from string import *
from random import shuffle

letters = "ABCDEFGHIJKLMNOPQRSTUVWXYZabcdefghijklmnopqrstuvwxyz"
text = "Welcome to secrecy, dear agents!"
def mono_cipher(pool=None):
    #generate the mono cipher
    if pool is None:
        pool = letters + digits 
        pool_original = list(pool)
        pool_shuffle  = list(pool)
        shuffle(pool_shuffle)
        return dict(zip(pool_original, pool_shuffle))
    
def inverse_monocipher(monoalpha_cipher):
    #return the inverse given the monocipher (dict)
    inverse_mono = {}
    #for loop for the key values in cipher 
    for key, value in monoalpha_cipher.items():
        inverse_mono[value] = key
        return inverse_mono
    
def encryption(text, monoalpha_cipher):
    encrypt_text = []
    #for loop on the letters inside the text message \n",
    for letter in text:
        encrypt_text.append(monoalpha_cipher.get(letter, letter))
        return " ".join(encrypt_text)
    
def decryption(encrypt_text, monoalpha_cipher):
    return encryption(encrypt_text, inverse_monocipher(monoalpha_cipher))

# generates the cipher 
cipher = mono_cipher()
# prints the cipher 
print(cipher)
# encrypts the given text from generated cipher 
encrypt_text = encryption(text, cipher)
print(encrypt_text)
# decrypts the encrypted text with generated cipher 
decryp = decryption(encrypt_text, cipher) 
print(decryp)
```

```python
# index of elements stored in dictionaries 
d1 = {'A': 0, 'B': 1, 'C': 2, 'D': 3, 'E': 4,'F': 5,
      'G': 6, 'H': 7, 'I': 8, 'J': 9, 'K': 10,
      'L': 11, 'M': 12, 'N': 13, 'O': 14, 'P': 15,
      'Q': 16, 'R': 17, 'S': 18, 'T': 19,'U': 20, 'V': 21,
      'W': 22, 'X': 23, 'Y': 24, 'Z': 25}

d2 = {0: 'A', 1: 'B', 2: 'C', 3: 'D', 4: 'E',
      5: 'F', 6: 'G', 7: 'H', 8: 'I', 9: 'J',
     10: 'K', 11: 'L', 12: 'M', 13: 'N', 14: 'O',
     15: 'P', 16: 'Q', 17: 'R', 18: 'S', 19: 'T',
     20: 'U', 21: 'V', 22: 'W', 23: 'X', 24: 'Y', 25: 'Z'}

# defined function to generate key 
def generatekey(text, key):
    x = 0
    # while loop to generate new key 
    while True:
        if len(key) == len(text):
            break
        if text[x] == " ":
            x += 1 # increment it 
        else:
            key += text[x] # increment the text key element 
            x += 1
    return key

# Autokey encryption 
def encryption(text, newkey):
    encryp_txt = " "
    x = 0
    for letter in text:
        if letter == " ":
            encryp_txt += " "
        else:
            # formulate to encipher text from plaintext by addition
            y = (d1[letter] + d1[newkey[x]]) % 26
            x += 1 # increment it
            encryp_txt += d2[y]
    return encryp_txt

# Autokey decryption 
def decryption(encryp_txt, newkey):
    decryp_txt = " "
    x = 0 
    for letter in encryp_txt:
        if letter == " ":
            decryp_txt += " "
        else:
            # formulate to decipher text to original by substraction
            y = (d1[letter] - d1[newkey[x]]) % 26
            x += 1
            decryp_txt += d2[y]
    return decryp_txt

def main():
    text = "COMFY MATTRESS"
    key = "NICE"
    newkey = generatekey(text, key)
    encryp_txt = encryption(text, newkey)
    decryp_txt = decryption(encryp_txt, newkey)
    # output the enciphered and deciphered texts after display of plain text
    print("Plain text: " + text + "\n")
    print("Using autokey cipher ↓")
    print("The encrypted text:" + encryp_txt)
    print("The decrypted text:" + decryp_txt) 

# Executes main 
if __name__ == "__main__":
    main()
```
